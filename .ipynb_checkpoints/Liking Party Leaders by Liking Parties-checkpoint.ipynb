```python
def weighted_mean(x, **kws):
    val, weight = map(np.asarray, zip(*x))
    val, weight = val[~np.isnan(val)],weight[~np.isnan(val)]
#     raise Exception
    return (val * weight).sum() / weight.sum()
```

```python
def party_leader_pop(tup_list,title,xlabel,ylabel,y2label):

    plt.figure(figsize=(14,8));
    dk_axis_count=1
    for tup in tup_list:

        party = tup[0]
        pol = tup[1]
        pol_name = tup[2]
        col = tup[3]

        best_weight_variable = search(BES_Panel,"wt_(core|full|new|fresh)",
                                      mask=BES_Panel[party].notnull() ).sort_values(ascending=False).index[0]
        best_weight_series = BES_Panel[best_weight_variable]
        print(best_weight_variable)

        BES_Panel["y"] = list(zip(BES_Panel[pol].replace("Don't know",np.nan).cat.codes.replace(-1,np.nan),best_weight_series ))
        mask = BES_Panel[pol].notnull()&BES_Panel[party].notnull()&best_weight_series.notnull()
        ax = sns.lineplot(x=BES_Panel[party][mask],
                     y=BES_Panel["y"][mask],
                     label=pol_name,color=col,estimator=weighted_mean)


        BES_Panel["y"] = list(zip( BES_Panel[pol].apply(lambda x: float(x=="Don't know")*10 if pd.notnull(x) else np.nan),
                       best_weight_series ))
        ax = sns.lineplot(x=BES_Panel[party][mask],
                     y=BES_Panel["y"][mask],
                     label=pol_name+" (like DK)",color=col,estimator=weighted_mean)
        ax.lines[dk_axis_count].set_linestyle("--")
        dk_axis_count = dk_axis_count+2

    def CtoF(x):
        return x * 10

    def FtoC(x):
        return x / 10

    secaxy = ax.secondary_yaxis('right', functions=(CtoF, FtoC))
    secaxy.set_ylabel(y2label)

    plt.ylabel(ylabel);
    plt.xlabel(xlabel);
    plt.title(title,fontsize=20);
    _=plt.ylim([0,10]);
```

```python
%%time

tup_list = [("likeConW4","likeCameronW4","Cameron",'b',),
            ("likeConW11","likeMayW11","May",'y',),
            ("likeConW17","likeBorisW17","Johnson",'g',)]
title = "Pre General Election Campaigns:\nHow much people like the party leader by how much they like the Conservative Party"
xlabel = "How much do you like or dislike the Conservative Party?\nStrongly dislike (0) - Strongly like (10)"
ylabel = "How much do you like or dislike each of the following political figures?\nStrongly dislike (0) - Strongly like (10)"
y2label = "'like == Don't Know'%"

party_leader_pop(tup_list,title,xlabel,ylabel,y2label)
```
