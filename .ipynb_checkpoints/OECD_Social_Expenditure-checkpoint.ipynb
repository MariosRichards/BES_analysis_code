```python
OECD_data_files = BES_small_data_files + os.sep + "Other_datasets"+ os.sep + "OECD_Social_Expenditure" + os.sep
```

```python
SE_2011 = pd.read_csv(OECD_data_files+"IDD-share-transfers-2011.csv")
SE_2014 = pd.read_csv(OECD_data_files+"IDD-share-transfers-2014.csv")
```

```python
set(SE_2011["States"].values).symmetric_difference(set(SE_2014["States"].values))
# Latvia in 2014 dataset, but not 2011

SE_2014 = SE_2014.drop( SE_2014[SE_2014["States"]=="Latvia"].index ).reset_index(drop=True)

# Quintiles by Equivalised Disposable Income
```

```python
# make sure all shares add up to 100%
all_quintiles = ["Q1","Q2","Q3","Q4","Q5"]

SE_2011[all_quintiles] = SE_2011[all_quintiles].apply(lambda x: x/SE_2011[all_quintiles].sum(axis=1))
SE_2014[all_quintiles] = SE_2014[all_quintiles].apply(lambda x: x/SE_2014[all_quintiles].sum(axis=1))
```

```python
SE_sum = SE_2014.copy()

SE_sum[all_quintiles] = (SE_2014[all_quintiles] + SE_2011[all_quintiles])
SE_sum["Q5-Q1"] = SE_sum["Q5"]-SE_sum["Q1"]

sorted_States = SE_sum.sort_values(by="Q5-Q1",ascending=True)["States"]
# "Q1","Q2","Q3","Q4","Q5"
```

```python
# order all states by 2011 Q5 values

SE_2011["States"] = SE_2011["States"].astype('category').cat.reorder_categories(sorted_States, ordered=True)
SE_2014["States"] = SE_2014["States"].astype('category').cat.reorder_categories(sorted_States, ordered=True)
```

```python
# SE_2011.drop([],axis=1).sort_values(by="States",ascending=True)
```

```python
SE_flat = SE_2011.drop([],axis=1).sort_values(by="States",ascending=True).copy()

SE_flat.index = SE_flat["States"]
SE_flat = SE_flat.drop("States",axis=1).stack().reset_index()
SE_flat.columns = ["States","Quintile","Share"]
SE_flat

f, ax = plt.subplots(figsize=(12,8))

ax = sns.pointplot(ax=ax, x="States", y="Share", hue="Quintile", data=SE_flat)
plt.xticks(rotation=90)
# ax.setp(SE_2011_flat["States"].unique(), rotation=90)
```
