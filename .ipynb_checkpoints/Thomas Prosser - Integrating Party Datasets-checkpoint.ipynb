```python
def get_dta_header_labels(filepath):

    reader = pd.io.stata.StataReader( filepath )
    header = reader.variable_labels()
    labels = reader.value_labels()
    return header,labels

def not_zero_or_nan(x):
    if pd.isnull(x):
        return False
    elif x==0:
        return False
    else:
        return True
```
