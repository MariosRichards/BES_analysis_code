```python
wave = "W13_only"

BES_numeric = pd.read_hdf( BES_data_folder+"BESnumeric"+wave+".hdf", "BESnumeric"+wave )
print("BES_numeric",  BES_numeric.shape )
var_type    = pd.read_hdf( BES_data_folder+"var_type"+wave+".hdf", "var_type" )
print("var_type",  var_type.shape )
BES_reduced = pd.read_hdf( BES_data_folder+"BES_reduced"+wave+".hdf", "BES_reduced"+wave )
print("BES_reduced",  BES_reduced.shape )

fname = BES_data_folder+"cat_dictionary"+wave+".pkl"
with open(fname, "rb") as f:
    cat_dictionary = pickle.load( f )
```

```python
# Okay - it gets confusing if you look at Scotland/Wales
# keep it simple, look at England alone

England = (BES_numeric["country"]==0.0)
```

```python
Lab2015 = BES_reduced.profile_past_vote_2015_Labour.astype('bool')
Lab2017 = BES_reduced.generalElectionVote_Labour.astype('bool')
voted2015 = BES_reduced["profile_turnout_2015_Yes, voted"].astype('bool')
voted2017 = BES_reduced["genElecTurnoutRetro_Yes, voted"].astype('bool')

used_vars = ["profile_past_vote_2015_Labour",
             "generalElectionVote_Labour",
             "profile_turnout_2015_Yes, voted",
             "genElecTurnoutRetro_Yes, voted"]
```

```python
# how do we feel about non-voters?
# should they be included in gained/lost?


voting_path = BES_reduced.generalElectionVote_Labour.astype('category')
voting_path.cat.set_categories(["kept","gained","lost","never had"],inplace=True)
voting_path[Lab2015  &  Lab2017 & voted2015 & voted2017]  = "kept"
voting_path[~Lab2015 &  Lab2017 & voted2015 & voted2017]  = "gained"
voting_path[Lab2015  & ~Lab2017 & voted2015 & voted2017]  = "lost"
voting_path[~Lab2015 & ~Lab2017 & voted2015 & voted2017]  = "never had"
voting_path.name = "Labour 2015 -> Labour 2017"
```

```python
kept   = (voting_path == "kept")
gained = (voting_path == "gained")
lost   = (voting_path == "lost")
never  = (voting_path == "never had")
voting_path.value_counts()
```

```python
def logit_base_predict(base_group, base_group_str,
                       predict_group, predict_group_str,
                       area):
    
    train_cols = BES_reduced[(base_group|predict_group) & area]\
        .drop([x for x in BES_reduced.columns if ("generalElectionVote" in x) | ("profile_past_vote_2015" in x)],axis=1)\
        .drop(used_vars,axis=1, errors='ignore')\
        .drop([x for x in BES_reduced.columns if "gor" in x],axis=1)
     # indirect heavy correlation with whether you voted Lab2015!


    tol = 1
    Q, R = np.linalg.qr(train_cols)
    not_independent = np.where(np.abs(R.diagonal()) <= tol)[0]
    print( "Dropping these variables because they aren't independent:\n" ,train_cols.columns[not_independent].values )
    train_cols = train_cols.drop(train_cols.columns[not_independent],axis=1)


    target = (voting_path[(base_group|predict_group) & area] == predict_group_str)



    logit = sm.Logit( target,
                      train_cols )

    # fit the model

    result = logit.fit(method='bfgs',maxiter=1000) 
    print(base_group_str," ---> ",predict_group_str)
    display(result.params[result.pvalues<0.05].sort_values())
```

```python
# kept vs never had
logit_base_predict(base_group = kept, base_group_str = "kept",
                   predict_group = never, predict_group_str = "never had",
                   area = England)
```

```python
# kept vs gained
logit_base_predict(base_group = kept, base_group_str = "kept",
                   predict_group = gained, predict_group_str = "gained",
                   area = England)
```

```python
# kept vs lost
logit_base_predict(base_group = kept, base_group_str = "kept",
                   predict_group = lost, predict_group_str = "lost",
                   area = England)
```

```python
# gained vs lost
logit_base_predict(base_group = gained, base_group_str = "gained",
                   predict_group = lost, predict_group_str = "lost",
                   area = England)
```
