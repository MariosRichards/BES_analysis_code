```python
links = []

# 2005 -> 2010, weighted
first_elec  = "2005"
second_elec = "2010"
crosstab = pd.crosstab(index   = BES_Panel[ 'profile_past_vote_'+first_elec  ],
                       columns = BES_Panel[ 'profile_past_vote_'+second_elec ],
                       values  = BES_Panel[ "wt_full_W8" ],
                       aggfunc = sum,
                       normalize=True)*100

for col in crosstab.columns:
    second_party = replace_list[ party_list.index( col ) ]
    for ind in crosstab[col].index:
        first_party = replace_list[ party_list.index( ind ) ]
        d = {'source': first_party+first_elec, 'target': second_party+second_elec, 'value': crosstab[col][ind]}
        links.append(d)
```

```python
w = SankeyWidget(links=links, margins=dict(top=0, bottom=0, left=70, right=70))
w
```
