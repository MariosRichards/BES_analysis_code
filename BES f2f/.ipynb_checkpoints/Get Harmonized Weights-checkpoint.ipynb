```python
def weighted_value_counts(x, wts, *args, **kwargs):
    normalize = kwargs.get('normalize', False)
    perc = kwargs.get('perc', False)
    decimal_places = kwargs.get('decimal_places', 2)
    suppress_raw_samplesize = kwargs.get('suppress_raw_samplesize', False)
    
    ascending = kwargs.get('ascending', True)
    c0 = x.name 
    c1 = wts.name
    df = pd.concat([x,wts],axis=1)
    xtmp = df.groupby(c0).agg({c1:'sum'}).sort_values(c1,ascending=False)
    s = pd.Series(index=xtmp.index, data=xtmp[c1], name=c0)
    s.name = "weighted_sample_size"
    if normalize:
        s = s / df[c1].sum()
        s.name = "weighted_sample_fraction"
    if normalize and perc:
        s = s*100
        s.name = "weighted_sample_percentage"
    s = s.round(decimal_places)
    if decimal_places==0:
        s=s.astype('int')
        
    if not suppress_raw_samplesize:
        output = pd.DataFrame([s,x[wts.notnull()].value_counts()]).T
        output.columns = [s.name,"raw_sample_size"]
        output.index.name = x.name
        output.sort_values(by=s.name,inplace=True, ascending=ascending)
    else:
        output = s
    return output

def get_age_cat(x):
    if x<18:
        return "<18"
    elif x>=18 and x<=24:
        return "18-24"
    elif x>=25 and x<=34:
        return "25-34"
    elif x>=35 and x<=44:
        return "35-44"
    elif x>=45 and x<=54:
        return "45-54"
    elif x>=55 and x<=64:
        return "55-64"
    elif x>=65 and x<=75:
        return "65-75"
    elif x>75:
        return "75+"
    
def get_inc_quint(x):
    labels = ["1st_inc_quint","2nd_inc_quint","3rd_inc_quint","4th_inc_quint","5th_inc_quint",]
    return pd.qcut(x,5,labels).astype(str)

def get_demo_cat(x):
    gender_replace_dict = {1.0:"Female",0.0:"Male"}
    return gender_replace_dict[x["gender_complete"]]+"|"+get_age_cat(x["age"]) +"|"+x["anyQuals"]+"|"+region.loc[x["uid"],"region"]
#+"|"+x["inc_quints"] 

def how_close_to_actual_wt_combo_with_diffs(x,actual_results,demo_cat_var_list,selection_weights,mask,reg_factor,
                                            target_list = ["vote_complete"],drop_list = ["OTH"]
                                 
                                 
                                ):
    
#     actual_results = args[0]
#     demo_cat_var_list = args[1]
#     selection_weights = args[2]
    
#     print(selection_weights.sum())
#     demo_cat_var_list = ["gender_complete","age_gp","anyQuals","region"]
    total_wt_max = 5
    total_wt_min =.2

    wt_dict = f2f_harmonised[demo_cat_var_list][mask].apply(lambda x: dict(zip(x.dropna().unique(),
                                                                               np.ones(len(x.dropna().unique())))))
    count = 0
    for col in wt_dict.keys():
        for cat in wt_dict[col].keys():
#             wt_dict[col][cat]=np.log10(x[count])
            wt_dict[col][cat]=10**(x[count])
            count=count+1

    wt_array = f2f_harmonised.loc[mask,demo_cat_var_list].apply(lambda x: x.replace(wt_dict[x.name]))
    if selection_weights is not None:
        wt_array["sel_wt"] = selection_weights
    # multiply weights together
    f2f_harmonised.loc[mask,"temp_wts"] = wt_array.apply(np.prod,axis=1)
    # first normalisation
    f2f_harmonised.loc[mask,"temp_wts"] = f2f_harmonised.loc[mask,"temp_wts"]/f2f_harmonised.loc[mask,"temp_wts"].mean()
    # cap weights at 1/5x - 5x
    f2f_harmonised.loc[mask,"temp_wts"] = f2f_harmonised.loc[mask,"temp_wts"].apply(lambda x: min(max(x,total_wt_min),total_wt_max))
    # second normalization
#     print(f2f_harmonised.loc[mask,"temp_wts"].min(),f2f_harmonised.loc[mask,"temp_wts"].max(),)
    f2f_harmonised.loc[mask,"temp_wts"] = f2f_harmonised.loc[mask,"temp_wts"]/f2f_harmonised.loc[mask,"temp_wts"].mean()
#     print(f2f_harmonised.loc[mask,"temp_wts"].min(),f2f_harmonised.loc[mask,"temp_wts"].max(),)
    
#     print("in_func max,min",f2f_harmonised.loc[mask,"temp_wts"].max(),f2f_harmonised.loc[mask,"temp_wts"].min())   
    diffs = pd.Series()
    for target in target_list:
        pred = weighted_value_counts(f2f_harmonised[target][mask],
                                       f2f_harmonised["temp_wts"][mask],normalize=True,perc=True)["weighted_sample_percentage"]\
            .drop(drop_list,errors='ignore')
        diffs = diffs.append(pred-pd.Series(actual_results).loc[pred.index])
#         print(target, pred, diffs)
#         raise Exception
        
#     print(diffs.values)
#     return diffs.apply(lambda x: x*x).sum()+(pd.Series(x).apply(lambda x: np.log10(x)**2).sum())
    regularisation_cost = (pd.Series(x).apply(lambda x: x**2).sum())*reg_factor
    return diffs.apply(lambda x: x*x).sum()+regularisation_cost, diffs

def how_close_to_actual_wt_combo(x,actual_results,demo_cat_var_list,selection_weights,mask,reg_factor,target_list = ["vote_complete"],drop_list = ["OTH"]):
    cost, diffs = how_close_to_actual_wt_combo_with_diffs(x,actual_results,demo_cat_var_list,selection_weights,mask,reg_factor,
                                                         target_list,drop_list)
    return cost

```
