```python
def create_closest_map(var1, var2, dataset):
    ct = (pd.crosstab(BES_df_list[dataset][var1],BES_df_list[dataset][var2],normalize='index').round(2)*100).astype(int)
    replace_dict = ct.idxmax(axis=1).to_dict()
    # only compare elements present in both
    mask = BES_df_list[dataset][var1].notnull()&BES_df_list[dataset][var2].notnull()
    accuracy = (BES_df_list[dataset][var1][mask].replace(replace_dict)==BES_df_list[dataset][var2][mask].astype('object')).mean()
    return replace_dict, accuracy, (BES_df_list[dataset][var1].isnull().sum()/BES_df_list[dataset][var1].shape[0], 
        BES_df_list[dataset][var2].isnull().sum()/BES_df_list[dataset][var1].shape[0], )
```

```python
def closest_map_matrix(var_list, dataset):
    df = pd.DataFrame()
    for var1 in var_list:
        for var2 in var_list:
            a,b,c = create_closest_map(var1, var2, dataset)
            df.loc[var2,var1] = b
    df.index.name ='to'
    df.columns.name ='from'            
    return df.round(2)
```
