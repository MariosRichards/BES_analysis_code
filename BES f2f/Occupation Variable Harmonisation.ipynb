```python
def create_closest_map(var1, var2, dataset,BES_or_harmonised="BES", nan_list = []):
    if BES_or_harmonised=="BES":
        ser1 = BES_df_list[dataset][var1].replace({k:np.nan for k in nan_list})
        ser2 = BES_df_list[dataset][var2].replace({k:np.nan for k in nan_list})
    elif BES_or_harmonised=="f2f_harm":
        ser1 = f2f_harmonised.loc[f2f_harmonised["dataset"]==dataset,var1].replace({k:np.nan for k in nan_list})
        ser2 = f2f_harmonised.loc[f2f_harmonised["dataset"]==dataset,var2].replace({k:np.nan for k in nan_list})
    else:
        raise Exception("BES_or_harmonised != 'BES' or 'f2f_harm'")
        
    ct = (pd.crosstab(ser1,ser2,normalize='index').round(2)*100).astype(int)
    replace_dict = ct.idxmax(axis=1).to_dict()
    # only compare elements present in both
    mask = ser1.notnull()&ser2.notnull()
    accuracy = (ser1[mask].replace(replace_dict)==ser2[mask].astype('object')).mean()
    return replace_dict, accuracy, (ser1.isnull().sum()/ser2.shape[0], 
        ser2.isnull().sum()/ser1.shape[0], )
```

```python
def closest_map_matrix(var_list, dataset, BES_or_harmonised="BES", nan_list = []):
    df = pd.DataFrame()
    for var1 in var_list:
        for var2 in var_list:
            a,b,c = create_closest_map(var1, var2, dataset,BES_or_harmonised,nan_list)
            df.loc[var2,var1] = b
    df.index.name ='to'
    df.columns.name ='from'            
    return df.round(2)
```

```python
focal_vars = ["rghclass"]
n_components = 4
def stringy_value_counts(ser):
    vc = ser.value_counts(normalize=True)
    vc.index = [str(x) for x in vc.index]
    return vc
    
def find_related_variable(dataset,ser,n_components):

    stub_not_cat = ~BES_df_list[dataset].dtypes.apply(lambda x: x.name).isin(["category","object"])
    stub_not_cat_list = stub_not_cat[stub_not_cat].index

    pca_comps = PCA(n_components=n_components).fit_transform(pd.get_dummies(ser))
    for comp in range(n_components):
        ser = BES_df_list[dataset]
        object_dtypes = ser.select_dtypes('object').columns
        ser[object_dtypes] = ser[object_dtypes].astype('category')
        pca0_corrs = pd.get_dummies(BES_df_list[dataset],prefix_sep='|').corrwith(pd.Series(pca_comps[:,comp],
                                                                             index=BES_df_list[dataset].index))
        temp = pd.DataFrame(pca0_corrs.dropna().sort_values()).reset_index()
        
        temp["stub"] = temp["index"].apply(lambda x: "|".join(x.split("|")[:-1]) if x not in stub_not_cat_list else x)
        temp["cat"] = temp["index"].apply(lambda x: x.split("|")[-1]  if x not in stub_not_cat_list else "")
        # temp.apply(lambda x: BES_df_list[dataset][x["stub"]],axis=1)
#         temp[temp["stub"]==""] # not a categorical variable
#         not_cat_var = temp["stub"]==""
#         temp.loc[not_cat_var,"stub"] = temp.loc[not_cat_var,"cat"]
#         temp.loc[not_cat_var,"cat"] = ""
        not_cat_var = temp["cat"]==""
        temp["fraction"] = 1.0
        temp.loc[~not_cat_var,"fraction"] = temp[~not_cat_var].apply(lambda x: stringy_value_counts(BES_df_list[dataset][x["stub"]]).loc[str(x["cat"])],axis=1)
        temp["abs_corr"] = temp[0].abs()
        if comp==0:
            temp2 = temp.groupby("stub").apply(lambda x: sum(x["fraction"]*x["abs_corr"]))
        else:
            temp2 = temp2 + temp.groupby("stub").apply(lambda x: sum(x["fraction"]*x["abs_corr"]))
    return temp2

# temp2 = find_related_variable(dataset,focal_vars,n_components)

def filter_related_variables(temp2, dataset, threshold=.4, drop_str=None):

    ser = temp2[temp2>threshold]
    ser = ser.reset_index()
    if drop_str is not None:
        mask = ser["stub"].apply(lambda x: re.search(drop_str,x) is None)
        ser = ser[mask]
    ser["label"] = ser.apply(lambda x: BES_label_list[dataset][x["stub"]],axis=1)            
    return ser.sort_values(by=0)
#     temp2.sort_values()
```
