```python
variable_descr_dict = {}
variable_descr_no = {}

header_list = [header1,header2,header3,header4,header5]

header_to_vardescr = {0:"var_descr_A",
                      1:"var_descr_B",
                      2:"var_descr_C",
                      3:"var_descr_D",
                      4:"var_descr_E",}

for header_no,var_descr in header_to_vardescr.items():
    header = header_list[header_no]
    for k,v in header.items():
        if k in Int_variable_df_fi[var_descr].values:
            continue
        if v not in variable_descr_dict.keys():
            variable_descr_dict[v] = [k]
            variable_descr_no[v] = 1
        else:
            variable_descr_dict[v].append(k)
            variable_descr_no[v] = variable_descr_no[v] + 1
#     {v:k for k,v in header.items() if k not in Int_variable_df_fi[var_descr].values}
```
