```python
def get_date(x):
    if pd.notnull(x):
        return str(int(x))
    else:
        return "nan"

missing_narrow = ["99. MISSING","9999. MISSING",'96. [SEE ELECTION STUDY NOTES]']
missing_narrow_dict = dict(zip(missing_narrow,[np.nan]*len(missing_narrow))) 

# '99. MISSING'
missing = ["96. NOT APPLICABLE: NO SECOND ROUND","99. MISSING","9996. NOT APPLICABLE: NO SECOND ROUND","9999. MISSING","96. [SEE ELECTION STUDY NOTES]"]
missing_dict = dict(zip(missing,[np.nan]*len(missing)))


def is_missing(x):
    return (x == "99. MISSING")|(x=="9999. MISSING")|(x=="96. [SEE ELECTION STUDY NOTES]")
    

def create_date_column(base_int):
    
    d_str,m_str,y_str = tuple(base_date_dict[base_int])
    date_temp = CSES_int_final[[base_int+d_str,base_int+m_str,base_int+y_str]]
    #.replace(date_replace_dict).replace(missing_dict).astype('object')
    
    # only occurs in IMD1012 and is consistent for all date compoents
    non_applicable = date_temp[[base_int+d_str,base_int+m_str,base_int+y_str]]\
        .apply(lambda x: (x == "96. NOT APPLICABLE: NO SECOND ROUND")|(x=="9996. NOT APPLICABLE: NO SECOND ROUND")).any(axis=1)
    # missing values appear only in IMD1013 and IMD5024
    # no overlap - but script will break if you add a variable that does have overlap!

    
    year_and_month_present_missing_day = date_temp[base_int+d_str].apply(lambda x: is_missing(x))&\
                                        date_temp[base_int+m_str].apply(lambda x: not is_missing(x))&\
                                        date_temp[base_int+y_str].apply(lambda x: not is_missing(x))
    
    date_temp[base_int+d_str][year_and_month_present_missing_day]==18
    
    year_or_month_missing = date_temp[base_int+m_str].apply(lambda x: is_missing(x))|\
                            date_temp[base_int+y_str].apply(lambda x: is_missing(x))
    
    date_temp = date_temp.replace(date_replace_dict).replace(missing_dict).astype('object')    
    

#    date_temp = date_temp.replace(date_replace_dict).replace({"96. NOT APPLICABLE: NO SECOND ROUND","9996. NOT APPLICABLE: NO SECOND ROUND"])
    
#     # replace missing day with random day from 1-30
#     date_temp[base_int+d_str]= date_temp[base_int+d_str].apply(lambda x: x if pd.notnull(x) else int(np.ceil(np.random.rand())*30)  )

#         # replace missing day with random day from 1-30
#     date_temp[base_int+m_str]= date_temp[base_int+m_str].apply(lambda x: x if pd.notnull(x) else int(np.ceil(np.random.rand())*12)  )
    
    # if month and year present, but day missing -> replace with 18th (highest frequency, mid month, better than trying to recretate the exact dist)
    # if month or year not present -> replace with 25 days after election
    
    # specify which of these holds in non-resp
    
    # you need distinct handling for cases where it's legitimately missing
#     ["96. NOT APPLICABLE: NO SECOND ROUND","99. MISSING","9996. NOT APPLICABLE: NO SECOND ROUND","9999. MISSING"]

    
    null_date_rows = date_temp.isnull().any(axis=1)
    date = date_temp[base_int+d_str].apply(get_date).astype(str)+"\\"+\
        date_temp[base_int+m_str].apply(get_date).astype(str)+"\\"+\
        date_temp[base_int+y_str].apply(get_date).astype(str)
    date[null_date_rows] = np.nan
    
    merged_date = pd.to_datetime(date, format = "%d\%m\%Y")
    if base_int=="IMD1013": ## impute using election day
        merged_date[year_or_month_missing] = election_dates.loc[CSES_int_final["IMD1003"]][year_or_month_missing.values]+\
                                                pd.Timedelta('25 days')
    
    non_resp = merged_date.copy()
    non_resp = np.nan
    non_resp= pd.Series(index=  merged_date.index)
    non_resp[:] =np.nan
    non_resp[year_and_month_present_missing_day] = "year_and_month_present_missing_day"
    if base_int=="IMD1013": ## impute using election day
        non_resp[year_or_month_missing] = "year_or_month_missing"
    else:
        non_resp[year_or_month_missing] = "missing"
    
    non_resp[non_applicable] = "non_applicable"
    non_resp = non_resp.astype('category')
    
    
    return merged_date, non_resp
    

# IMD1011_X
# DATE 1ST ROUND ELECTION BEGAN
base_int = "IMD1011"
IMD1011_date,IMD1011_date_non_resp = create_date_column(base_int)
election_dates = IMD1011_date.groupby(CSES_int_final["IMD1003"]).first()

# IMD1012_X
# DATE 2ND ROUND ELECTION BEGAN
base_int = "IMD1012"
IMD1012_date,IMD1012_date_non_resp = create_date_column(base_int)

# DATE QUESTIONNAIRE ADMINISTERED
base_int = "IMD1013"
IMD1013_date,IMD1013_date_non_resp = create_date_column(base_int)

# IMD5024_X
# DATE ELECTION SCHEDULED
base_int = "IMD5024"
IMD5024_date,IMD5024_date_non_resp = create_date_column(base_int)

# IMD5025_X
# DATE ELECTION HELD
base_int = "IMD5025"
IMD5025_date,IMD5025_date_non_resp = create_date_column(base_int)
```

```python
# IMD1011_X
# DATE 1ST ROUND ELECTION BEGAN

# IMD1012_X
# DATE 2ND ROUND ELECTION BEGAN

# IMD1013_X
# DATE QUESTIONNAIRE ADMINISTERED

# IMD5024_X
# DATE ELECTION SCHEDULED

# IMD5025_X
# DATE ELECTION HELD

dmy_sub = ["_D","_M","_Y"]
num_sub = ["_2","_1","_3"]

flatten = lambda l: [item for sublist in l for item in sublist]
base_date_dict = {"IMD1011":dmy_sub,"IMD1012":dmy_sub,"IMD1013":dmy_sub,"IMD5024":num_sub,"IMD5025":num_sub}
base_date_dict_names = {"IMD1011":"DATE 1ST ROUND ELECTION BEGAN","IMD1012":"DATE 2ND ROUND ELECTION BEGAN",
                        "IMD1013":"DATE QUESTIONNAIRE ADMINISTERED","IMD5024":"DATE ELECTION SCHEDULED",
                        "IMD5025":"DATE ELECTION HELD"}


def get_date(x):
    if pd.notnull(x):
        return str(int(x))
    else:
        return "nan"
    
def create_date_column(ser):
    d_str,m_str,y_str = tuple(base_date_dict[ser.name])
    
    date_temp = input_df[[ser.name+d_str,ser.name+m_str,ser.name+y_str]].copy()
    date_temp_non_resp = date_temp.apply(lambda x: x if not isnum(x) and re.match(pattern,x) else np.nan).astype('category')
    
    null_date_rows = input_df[[ser.name+d_str,ser.name+m_str,ser.name+y_str]].isnull().any(axis=1)
    date = input_df[ser.name+d_str].apply(get_date)+"\\"+input_df[ser.name+m_str].apply(get_date)+"\\"+input_df[ser.name+y_str].apply(get_date)
    date[null_date_rows] = np.nan
    
    
    
    non_resp = ser.apply(lambda x: x if not isnum(x) and re.match(pattern,x) else np.nan).astype('category').copy()    
    non_resp.name = ser.name+"_DK"
    non_resp_cont_inf = non_resp.value_counts().sort_index().to_dict()    
    
    
    return pd.to_datetime(date, format = "%d\%m\%Y")    
    
def add_manifest_entry(resp_name,variable,response_type,resp_rows,non_resp_rows,resp_cont_inf,variable_type):
    # response_type = "response"|"nonresponse"
    #
    # get next manifest index
    if process_variable_manifest.shape[0]==0:
        man_ind = 0
    else:
        man_ind = np.max(process_variable_manifest.index)+1
    if variable_type=="date":
        process_variable_manifest.loc[man_ind] = [resp_name,
                                                  [resp_name+x for x in base_date_dict[resp_name]],
                                                  base_date_dict_names[resp_name],
                                                  response_type,
                                                  resp_rows,
                                                  non_resp_rows,
                                                  resp_cont_inf,
                                                  variable_type,
                                                  variable_manifest.set_index("index").loc[variable,"variable structural type"],

                                                 ]            
    else:
        process_variable_manifest.loc[man_ind] = [resp_name,
                                                  variable,
                                                  variable_manifest.set_index("index").loc[variable,"descr"],
                                                  response_type,
                                                  resp_rows,
                                                  non_resp_rows,
                                                  resp_cont_inf,
                                                  variable_type,
                                                  variable_manifest.set_index("index").loc[variable,"variable structural type"],

                                                 ]    
```

```python
pattern = "(9{2,8}\d?\.(00)?\.?)|(.*\[SEE ELECTION STUDY NOTES\])|(.*DON'T KNOW)|(.*REFUSED)|(.*NOT APPLICABLE)|(.*MISSING)|(.*Transition Periods)"

seminumeric_replace_dict = {"10. Democratic":10,"-10. AUTOCRATIC":-10,"00.00 THERE IS NO THRESHOLD":0,
                           "00. LEFT":0.0,"10. RIGHT":10.0}

date_replace_dict = {'01. JANUARY':1, '02. FEBRUARY':2, '03. MARCH':3, '04. APRIL':4, '05. MAY':5, '06. JUNE':6,
                     '07. JULY':7, '08. AUGUST':8, '09. SEPTEMBER':9, '10. OCTOBER':10, '11. NOVEMBER':11, '12. DECEMBER':12}

def isnum(x):
    return str(x).replace('.','',1).replace('-','',1).isdigit()

def list_categories(ser):
    return "|".join([str(x) for x in ser.cat.categories])


def process_var(ser,var_type):
    non_resp = ser.apply(lambda x: x if not isnum(x) and re.match(pattern,x) else np.nan).astype('category').copy()    
    non_resp.name = ser.name+"_DK"
    non_resp_cont_inf = non_resp.value_counts().sort_index().to_dict()
    
    if var_type == "seminumeric":
        
        resp=ser.copy()
        
        if ser.dtype.name=='object':
            resp = resp.replace(seminumeric_replace_dict)
        # FIX LATER
        
        resp = resp.apply(lambda x: x if isnum(x) else np.nan).astype('float32').copy()
        
        assert (non_resp.notnull().sum()+resp.notnull().sum()) == ser.shape[0]
        
        resp.name = ser.name+"_NUM"
        resp_cont_inf = (resp.min(),resp.max())
        
        
        return non_resp,resp,non_resp_cont_inf,resp_cont_inf
    
    elif var_type == "ordinal":
        
        resp = ser.astype('category').cat.remove_categories(non_resp.cat.categories)
        resp_cont_inf = dict(zip(resp.cat.categories,
                             range(0,len(resp.cat.categories))))
        
        resp = resp.cat.codes.replace(-1,np.nan).astype('float32')
        
        assert (non_resp.notnull().sum()+resp.notnull().sum()) == ser.shape[0]
        
        resp.name = ser.name+"_ORD"
        
        return non_resp,resp,non_resp_cont_inf,resp_cont_inf
    
    elif var_type == "dummy":
        
        resp = ser.astype('category').cat.remove_categories(non_resp.cat.categories)
        resp_cont_inf = resp.value_counts().sort_index().to_dict()
        
        resp = pd.get_dummies( resp, prefix = ser.name )
        resp.name = ser.name
        
        ## should we change this to numbers???
        
        assert (non_resp.notnull().sum()+resp.any(axis=1).sum()) == ser.shape[0]
        
#         resp.name = ser.name+"_DUM"
        
        
        return non_resp,resp,non_resp_cont_inf,resp_cont_inf    
    
    elif var_type == "party":
        
        ## change to replacing with partyfacts_id
        
#         resp = ser.astype('category').cat.remove_categories(non_resp.cat.categories).apply(lambda x: float(x[0:7]) if pd.notnull() else np.nan).astype('float32')
        resp = ser.apply(lambda x: cses_to_partyfacts_final.loc[x]).astype(float)
        resp_cont_inf = (resp.min(),resp.max())
        resp.name = ser.name+"_PTY"

    
        non_resp = ser.copy()
        non_resp[resp.notnull()]=np.nan
        #ser.apply(lambda x: x if not isnum(x) and re.match(pattern,x) else np.nan).astype('category').copy()    
        non_resp.name = ser.name
        non_resp.name = non_resp.name+"_DK"
        non_resp_cont_inf = non_resp.value_counts().sort_index().to_dict()    
        
        
        # what information would you really want here for party data?
        # should I change this with partyfacts_ids - go for min/max?
        #resp.value_counts().sort_index().to_dict()
        
#         resp.name = ser.name
        
        ## should we change this to numbers???
        
        assert (non_resp.notnull().sum()+resp.notnull().sum()) == ser.shape[0]
        

        
        
        return non_resp,resp,non_resp_cont_inf,resp_cont_inf
    
    elif var_type == "date":
        
        resp=ser.astype('category').cat.remove_categories(non_resp.cat.categories)
        
        if (resp.dtype.name=='object') or (resp.dtype.name=='category' and resp.cat.categories.dtype!='int64'):
            resp = resp.replace(date_replace_dict)
                
                
#         base_int = ser.name
        resp.name = ser.name.split("_")[0]   
                
        resp, non_resp = create_date_column(resp.astype('float32'))
        
        
        
        
        
        # what information would you really want here for party data?
        resp_cont_inf = resp.value_counts().sort_index().to_dict()
        
        
        
        ## should we change this to numbers???
        
        assert (non_resp.notnull().sum()+resp.notnull().sum()) == ser.shape[0]
        
        resp.name = ser.name+"_DAT"
        
        
        return non_resp,resp,non_resp_cont_inf,resp_cont_inf    
    
        ## some final processing here to get datetimes
        # maybe separate variables?
        
    elif var_type == "meta":
                        
        resp = ser.astype('category').cat.remove_categories(non_resp.cat.categories)
        
        # what information would you really want here for party data?
        resp_cont_inf = resp.value_counts().sort_index().to_dict()
        
#         resp = pd.get_dummies( resp, prefix = ser.name )
        resp.name = ser.name
        
        ## should we change this to numbers???
        
        assert (non_resp.notnull().sum()+resp.notnull().sum()) == ser.shape[0]
        
        resp.name = ser.name+"_MET"
        
        
        return non_resp,resp,non_resp_cont_inf,resp_cont_inf           

    
    # Date?    
```
