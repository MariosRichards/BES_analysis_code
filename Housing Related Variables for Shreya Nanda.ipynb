```python
def get_age_gp(x):
    if x<18:
        return "Under 18"
    if x<=25:
        return "18-25"
    if x<=35:
        return "26-35"
    if x<=45:
        return "36-45"
    if x<=55:
        return "46-55"
    if x<=65:
        return "56-65"
    if x>65:
        return "66+"
    else:
        return np.nan

BES_Panel["ageW8"].astype(float).apply(get_age_gp)
# .value_counts(dropna=False)
# BES_Panel["ageW8"].astype(float).apply(get_age_gp).notnull().sum()
```
