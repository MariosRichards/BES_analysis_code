```python
wave_list = list(range(1,21))
var_dict = {}
for wave in wave_list:
    
    data_subfolder = BES_data_folder + "W"+str(wave)+"_only" + os.sep
    
    if os.path.isfile(data_subfolder + "var_type.csv"):
        (var_type, cat_dictionary, new_old_col_names, old_new_col_names) = get_small_files(data_subfolder, encoding)
        var_dict[wave] = var_type.index
```

```python
new_waves = [21,22,23,24,25]
new_in_wave_dict = {}

for wave in new_waves:
    fname = BES_data_folder+"W"+str(wave)+"_only"+os.sep+"BES2019_W"+str(wave)+"_v25.0"
    BES_Panel = pd.read_stata(fname+".dta")
    BES_Panel.to_pickle(fname+'.zip',compression='zip')
    
    new_in_wave_dict[wave] = [x for x in BES_Panel.columns if x not in flatten(var_dict.values())]
    var_dict[wave] = BES_Panel.columns
```
