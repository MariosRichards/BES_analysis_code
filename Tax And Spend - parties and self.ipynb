```python
df = BES_Panel[search(BES_Panel,"taxSpend(Self|Con|Lab)W14").index].stack().reset_index()
df["id"] = df["id"].map(BES_Panel["wt_new_W14"])
df["level_1"] = df["level_1"].apply(lambda x: x.replace("taxSpend","").replace("W14",""))
df.columns = ["weights","ques_subject","value"]
ct = pd.crosstab(df["value"],df["ques_subject"],values=df["weights"],aggfunc=sum,normalize='columns')*100
df = ct.round(2).stack().reset_index()
df.columns = ["taxSpend","","Percentage"]
plt.figure(figsize=(14,8));
sns.barplot(data=df,x="taxSpend",y="Percentage",hue="");
import textwrap
plt.gca().set_xticklabels([textwrap.fill(e, 15) for e in BES_Panel["taxSpendSelfW14"].cat.categories]);
plt.title("Using the 0 to 10 scale below,\nwhere the end marked 0 means that government should cut taxes a lot and spend much less on health and social services,\nand the end marked 10 means that government should raise taxes a lot and spend much more on health and social services,\nwhere would you place yourself and the other parties on this scale?");
```
