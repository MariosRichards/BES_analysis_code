```python
import re, pickle


def process_xpath_dict(xpath_dict):
    triling_xpath_dict = {}
    for key in xpath_dict.keys():
        xpath_components = xpath_dict[key].split('/')[1:]
        triling_xpath_dict[key] = ""
        language_xpath_flag = False
        for xpath_comp in xpath_components:

            if re.search("@class='en-container'",xpath_comp):
                language_xpath_flag = True
            triling_xpath_dict[key] += "/"+xpath_comp
        if language_xpath_flag==True:
            triling_xpath_dict[key+"_es"] = ""
            triling_xpath_dict[key+"_fr"] = ""

            language_xpath_comp_flag = False
            # three possibilities
            # before language container (just copy straight)
            # after language container (copy and replace En -> Es/Fr)
            # language container (copy and replace @class='en-container' )
            for xpath_comp in xpath_components:
                
                if language_xpath_comp_flag:
#                     print(xpath_comp)
                    triling_xpath_dict[key+"_es"] += "/"+re.sub("En'", "Es'", xpath_comp)
                    triling_xpath_dict[key+"_fr"] += "/"+re.sub("En'", "Fr'", xpath_comp)
                elif re.search("@class='en-container'",xpath_comp):
                    triling_xpath_dict[key+"_es"] += "/"+re.sub("@class='en-container'", "@class='es-container'", xpath_comp)
                    triling_xpath_dict[key+"_fr"] += "/"+re.sub("@class='en-container'", "@class='fr-container'", xpath_comp)                    
                    language_xpath_comp_flag = True
                else:
                    triling_xpath_dict[key+"_es"] += "/"+xpath_comp
                    triling_xpath_dict[key+"_fr"] += "/"+xpath_comp
            
            triling_xpath_dict[key+"_en"] = triling_xpath_dict.pop(key)
    return(triling_xpath_dict)
```
